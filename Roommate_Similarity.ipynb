```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder


data = pd.read_excel('Dataset.xlsx')
df=data.copy()
#print(data)

course=input("Enter course:")
college=input("Enter the college:")
ea=input("Any extracurricular activities?")
sports=input("Interested sport:")
actor=input("Favourite actor/actress:")
major=input("Majors:")
hobby=input("Hobby:")
env=input("Environment:")
weekend=input("Weekend activity:")
music=input("Music type:")
sleep=input("Sleep pattern:")
diet=input("Diet:")
tidiness=input("Tidiness:")
cm=input("Conflict Management:")
uh=input("Undesirable habits:")
nat=input("Nature:")
ov=input("Occassional visitors?:")
sh=input("Study habits:")
sha=input("Sharing?:")
eq=input("Expected qualities:")

new_data = [course,college,ea,sports,actor,major,hobby,env,weekend,music,sleep,diet,tidiness,cm,uh,nat,ov,sh,sha,eq]
#df_new = pd.DataFrame(new_data)
#df_combined = data.append(df_new,index=data.columns, ignore_index=True)
data.loc[len(data)] = new_data
data.to_excel('modified.xlsx', index=False)
#print(data)

encoder = OneHotEncoder(sparse=False)
data_encoded = encoder.fit_transform(data)


cosine_similarities = cosine_similarity(data_encoded)


similarities_to_first_row = cosine_similarities[-1]
most_similar_indices = similarities_to_first_row.argsort()[:-6:-1]  


#print("Indices of top 5 most similar rows")
#print(most_similar_indices)
print("Top 5 most similar rows:")
for idx in most_similar_indices:
    print(data.iloc[idx])
```
